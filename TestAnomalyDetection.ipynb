```python
import face_recognition

#initialize anomaly_ct,correct_ct
anomaly_ct=0
correct_ct=0

#load images
known_image = face_recognition.load_image_file("Testkang0.jpg")

#jitters encodes the images by considering all angles of images, rotating zooming ,translating them etc
known_encodingS = face_recognition.face_encodings(known_image, num_jitters=100)
known_encoding=known_encodingS[0]


for i in range(0,9):
    imgName="Testkang"+str(i+1)+".jpg"
    unknown_image = face_recognition.load_image_file("Testkang"+str(i+1)+".jpg")
    #encode them
    #That will tell dlib to randomly distort your image 100 times (randomly zoomed, rotated, translated, flipped), 
    #take the encoding of each version of the image and return the average.
    #This can give you more generic face encodings that work better in some cases with the tradeoff that the algorithm will run more slowly.
    #Whether or not the increase in accuracy is worth the slower runtime depends on your use case.
    unknown_encodingS = face_recognition.face_encodings(unknown_image,num_jitters=81)


    known_face_encodings = [
    known_encoding
    ]

    known_face_names = [
        "Anoohya Bollampally"
    ]
    
    
    if(len(unknown_encodingS)==0):
        print("ANOMALY:no person or image not captured well"+str(i+1))
        anomaly_ct=anomaly_ct+1
    elif len(unknown_encodingS)>1:
        print("ANOMALY:multiple person "+str(i+1))
        anomaly_ct=anomaly_ct+1
    else:
        unknown_encoding=unknown_encodingS[0]
        
        #Using tolerance values lower than 0.6 will make the comparison more strict
        results=face_recognition.compare_faces([known_encoding],unknown_encoding,tolerance=0.63)
        if results[0]==True:
            status=objectDetectionAnomaly(imgName)
            if status==1:
                print("ANOMALY:External object detected"+str(i+1))
                anomaly_ct=anomaly_ct+1
            elif status==2:
                print("ANOMALY:multiple person "+str(i+1))
                anomaly_ct=anomaly_ct+1
            else:
                print("NO ANOMALY detected")
                correct_ct=correct_ct+1
        else:
            print("ANOMALY: different person detected "+str(i+1))
            anomaly_ct=anomaly_ct+1
            
#count of correctly classified images
#count of images having anomaly
print("Frames without anomaly:"+str(correct_ct))
print("Frames with anomaly:"+str(anomaly_ct))
probability=anomaly_ct/(anomaly_ct+correct_ct)
print(probability)
if(probability>0.441):
    print("probability of occurance of anomaly is more")
else:
    print("probability of occurance of anomaly is less")
```
